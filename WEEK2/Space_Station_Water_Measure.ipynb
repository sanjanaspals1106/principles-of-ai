```python
import heapq

def space_station_a_star(cap_x=7, cap_y=4, target=6):
    start = (0, 0)

    def heuristic(state):
        x, y = state
        return abs((x + y) - target) + min(abs(x - target), abs(y - target))

    def neighbors(state):
        x, y = state
        nbrs = []

        # Fill containers
        nbrs.append((cap_x, y))
        nbrs.append((x, cap_y))

        # Empty containers
        nbrs.append((0, y))
        nbrs.append((x, 0))

        # Transfer X → Y
        t = min(x, cap_y - y)
        nbrs.append((x - t, y + t))

        # Transfer Y → X
        t = min(y, cap_x - x)
        nbrs.append((x + t, y - t))

        return nbrs

    pq = []
    heapq.heappush(pq, (heuristic(start), 0, start, [start]))
    visited = set()

    while pq:
        f, g, state, path = heapq.heappop(pq)

        if state in visited:
            continue
        visited.add(state)

        x, y = state
        if x == target or y == target:
            return path, g

        for nxt in neighbors(state):
            if nxt not in visited:
                ng = g + 1
                nf = ng + heuristic(nxt)
                heapq.heappush(pq, (nf, ng, nxt, path + [nxt]))

    return None, None
```

```python
solution, steps = space_station_a_star()

if solution:
    print("Solution found using A* search:\n")
    for i, state in enumerate(solution):
        print(f"Step {i}: Container X = {state[0]}, Container Y = {state[1]}")
    print(f"\nTotal steps taken: {steps}")
else:
    print("No solution found.")
```
