```python
import heapq

CAP = (8, 5, 3)
START = (0, 0, 0)
TARGET = 4

def goal(state): return TARGET in state

def neighbors(s):
    a,b,c = s
    res = []
    for i,(x,cap) in enumerate(zip(s,CAP)):
        if x<cap:
            t=list(s); t[i]=cap; res.append(tuple(t))  # fill
        if x>0:
            t=list(s); t[i]=0; res.append(tuple(t))    # empty
        for j,(y,capj) in enumerate(zip(s,CAP)):
            if i!=j and x>0 and y<capj:
                t=list(s); transfer=min(x,capj-y); t[i]-=transfer; t[j]+=transfer; res.append(tuple(t))
    return res

def search(mode='bfs', max_depth=30):
    # f = g for BFS, f=-g for DFS
    pq = []
    heapq.heappush(pq,(0,0,START,[START]))
    visited = set()
    while pq:
        f,g,state,path = heapq.heappop(pq)
        if goal(state): return path
        if g>=max_depth or state in visited: continue
        visited.add(state)
        for n in neighbors(state):
            if n not in visited:
                new_g = g+1
                new_f = new_g if mode=='bfs' else -new_g
                heapq.heappush(pq,(new_f,new_g,n,path+[n]))
    return None

# ---------------- USAGE ----------------
for m in ['bfs','dfs']:
    print(f"\n{m.upper()} solution:")
    p=search(mode=m)
    for s in p: print(s)
```
