```python
class TrafficMDP:
    def __init__(self):
        self.states = [0, 1, 2]  # Low, Medium, High
        self.actions = ['SHORT', 'MEDIUM', 'LONG']
        self.rewards = [10, 2, -10]

    def transition(self, s, a):
        if a == 'LONG':
            return [(0.7, 1), (0.3, 2)] if s == 2 else \
                   [(0.6, 0), (0.4, 1)] if s == 1 else [(1, 0)]
        if a == 'SHORT':
            return [(0.6, 1), (0.4, 0)] if s == 0 else \
                   [(0.7, 2), (0.3, 1)] if s == 1 else [(1, 2)]
        return [(0.2, max(0, s-1)), (0.6, s), (0.2, min(2, s+1))]


def value_iteration(env, gamma=0.9, eps=0.01):
    V = [0] * 3

    while True:
        delta = 0
        for s in env.states:
            v = V[s]
            V[s] = max(
                sum(p * (env.rewards[s1] + gamma * V[s1])
                    for p, s1 in env.transition(s, a))
                for a in env.actions
            )
            delta = max(delta, abs(v - V[s]))
        if delta < eps:
            break

    return V


# Run
env = TrafficMDP()
V = value_iteration(env)

labels = ["Low", "Medium", "High"]

print("Optimal State Values:")
for i in range(3):
    print(f"{labels[i]} Traffic : {V[i]:.2f}")
```
