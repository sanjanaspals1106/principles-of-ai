```python
import random

class RobotMDP:
    def __init__(self):
        self.states = [0, 1, 2, 3]
        self.actions = ['PICK', 'MOVE', 'CHARGE']

    def step(self, s, a):
        if a == 'PICK':
            return (2, 8) if s == 1 else (s, -2)
        if a == 'MOVE':
            return (1, 5) if s == 2 else (s, -1)
        if a == 'CHARGE':
            return (1, 3) if s in [0, 3] else (s, -1)


def policy_iteration(env, gamma=0.9):
    V = {s: 0 for s in env.states}
    policy = {s: random.choice(env.actions) for s in env.states}

    while True:
        # Policy Evaluation
        while True:
            delta = 0
            for s in env.states:
                v = V[s]
                ns, r = env.step(s, policy[s])
                V[s] = r + gamma * V[ns]
                delta = max(delta, abs(v - V[s]))
            if delta < 0.01:
                break

        # Policy Improvement
        stable = True
        for s in env.states:
            old = policy[s]
            policy[s] = max(
                env.actions,
                key=lambda a: env.step(s, a)[1] +
                              gamma * V[env.step(s, a)[0]]
            )
            if old != policy[s]:
                stable = False

        if stable:
            break

    return policy


# Run
env = RobotMDP()
policy = policy_iteration(env)

labels = {
    0: "Idle-Low Battery",
    1: "Idle-Charged",
    2: "Carrying Load",
    3: "Charging Station"
}

print("Optimal Policy:\n")
for s in env.states:
    print(f"{labels[s]:20} → {policy[s]}")
```
